```python
from skimage import io
# Load an image from a file
photo = io.imread(r'image.jpg')
type(photo)  # Check the type of the loaded image

```

```python
plt.imshow(photo[::-1])
```

```python
plt.imshow(photo[:,::-1])  # Flipping the image horizontally
```

```python
plt.imshow(photo[10:100, 170:220])  # Cropping the image to a specific region
```

```python
plt.imshow(photo[::2, ::2])  # Display every second pixel in both dimensions
```

```python
plt.imshow(photo[:,:,0].T)
```

```python
import numpy as np
photo_masked = np.where(photo >  10, 20,0)  # Masking pixels with values greater than 200
plt.imshow(photo_masked)
```

```python
print(np.sum(photo))
print(np.prod(photo)) 
print(np.mean(photo))  # Mean value of the pixel intensities
print(np.std(photo))  # Standard deviation of the pixel intensities
print(np.var(photo))  # Minimum pixel intensity
print(np.argmax(photo))  # Maximum pixel intensity
print(np.argmin(photo))  # Index of the maximum pixel intensity
print(np.max(photo))  # Maximum pixel intensity
print(np.min(photo))  # Minimum pixel intensity
```
