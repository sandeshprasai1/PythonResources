```python
#matplotlib
import matplotlib.pyplot as plt
import numpy as np
labels=['grp1','grp2','grp3','grp4']
me=[20,35,33,24]
you=[43,66,54,70]

fig, ax=plt.subplots()

ax.bar(labels,me,label='Me')
ax.bar(labels,you,bottom=me,label='you')

ax.set_ylabel('age')
ax.set_xlabel('group member')
ax.set_title('gender ratio by group')
ax.legend()
plt.show()
```

```python
sub=['maths','phy','bio','chem']
sec1=[21,45,2,34]
sec2=[21,12,17,62]
sub_index=np.arange(len(sub))
width=0.2
plt.bar(sub_index,sec1,width,label='sec1 stds')
plt.bar(sub_index+width,sec2,width,label='sec2 stds')
plt.xticks(sub_index+width,sub)#xticks exchanges sub with index
```

```python
#lineplot
import numpy as num
x_axis=num.array([0,1,2,2,3])
plt.plot(x_axis,x_axis*2)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("y=2*x")
```

```python
#lineplot
import numpy as num
x_axis=num.array([0,1,2,2,3])
plt.plot(x_axis,x_axis**2,'green',label='y=x^2')
plt.legend()
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("y=x^2")
```

```python

y_axis = x_axis**2
empty_canvas = plt.figure()
create_axis = empty_canvas.add_axes([0, 0, 1, 1]) 

create_axis.plot(x_axis, y_axis, 'b')

create_axis = empty_canvas.add_axes([0, 0, 1, 1]) #left,bottom,width,height

create_axis.plot(x_axis, y_axis, 'b', label = "x = x**2")
create_axis.plot(x_axis, y_axis+(x_axis*3), 'g', label = "x = y+(x*3)")

create_axis.legend(loc=0)
```

```python
import matplotlib.pyplot as plt

data = {'Batch 1': 10, 'Batch 2': 15, 'Batch 3': 5, 'Batch 4': 20}
names = list(data.keys())
values = list(data.values())

fig, axs = plt.subplots(1,3, figsize=(10, 3))

axs[0].bar(names, values)
axs[1].scatter(names, values)
axs[2].plot(names, values)

fig.suptitle('Subplots')
```
